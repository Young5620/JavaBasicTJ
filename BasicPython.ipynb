```python
# 데이터 셋 분할
# 모델학습에 사용할 훈련(train)데이터와 모델 성능을 평가하는데 사용할 테스트(test)데이터를 분할한다.
# 보통 훈련 80% 테스트20%
# 전체 150개 데이터 중 테스트용 30개 훈련용 120개로 분류

from sklearn.model_selection import train_test_split
X_data = df . loc[ : , 'sepal_width' : 'petal_width']
y_data = df.loc[ : , 'target']

X_train , X_test , y_train , y_test = train_test_split(X_data, y_data, test_size= 0.2, shuffle=True, random_state=20) # 동일하게 랜덤으로 뽑아줌
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# KNN 알고리즘 

# K-Nearest-Neighbors K근접이웃
# K값에 따라 모델의 예측력이 달라지므로 적절한 K값을 설정해야한다.

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(X_train, y_train)
y_knn_pred = knn.predict(X_test)
knn_acc = accuracy_score(y_test,y_knn_pred)
knn_acc
# n=7 0.9666666666666667
# n=2 0.9
# n=1 0.9333333333333333
# n=20 1.0
# n=50 0.9333333333333333
# n=120 0.26666666666666666
```

```python
# SVM Support Vector Machine
#각 군집간의 거리(margin)를 최대한 멀리 유지하는 경계면을 찾는다.

from sklearn.svm import SVC
svc = SVC(kernel='rbf')
svc.fit(X_train, y_train)

y_svc_pred = svc.predict(X_test)
svc_acc = accuracy_score(y_test, y_svc_pred)
svc_acc
```

```python
# Logistic Regression

# 각 품종에 속하는 확률을 계산 - 1에 가까우면 그 클래스로 분류
#                                        - 0에 가까우면 아니라고 분류

from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression()
lrc.fit(X_train,y_train)
y_lrc_pred = lrc.predict(X_test)
lrc_acc = accuracy_score(y_test,y_lrc_pred)
lrc_acc
```

```python
# 의사결정나무 decision tree
# 트리의 각 분기점에는 데이터셋의 피처를 하나씩 위치시켜서 해당 피처에 대한 임의의 조건식을 가지고 계속 2개의 줄기로 가지를 나누면서 데이터를 구분

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth = 3 , random_state = 20)
dtc.fit(X_train,y_train)
y_dtc_pred = dtc.predict(X_test)
dtc_acc = accuracy_score(y_test , y_dtc_pred)
dtc_acc
```

```python
# 앙상블(보팅, 배깅)
# - 보팅(투표) : 하나의 모델 예측에 의존하는 것보다 여러 모델의 예측을 종합하면 예측력이 일반적으로 더 좋다.
# - 배깅 : 하나의 알고리즘을 여러번 돌린다.
from sklearn.ensemble import VotingClassifier
hvc = VotingClassifier(estimators = [('KNN',knn),('SVM',svc),('DT',dtc)],voting='hard')
hvc.fit(X_train,y_train)
y_hvc_pred = hvc.predict(X_test)
hvc_acc = accuracy_score(y_test, y_hvc_pred)
hvc_acc
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=50,max_depth=5,random_state=20)
rfc.fit(X_train,y_train)
y_rfc_pred = rfc.predict(X_test)
rfc_acc = accuracy_score(y_test, y_rfc_pred)
rfc_acc
```

```python
# 앙상블 부스팅
# XGBoost 모델은 캐글, 데이콘등 경진대회에 가장 많이 사용되는 알고리즘이다.
# 모델학습속도가 빠르고 예측력이 상당히 좋은편으로 알려져있다.
# 여러개의 모델을 순차적으로 학습한다. 오차를 줄일 수 있는 방향으로 모델을 계속 업데이트한다.
# 여러 모델을 동시에 학습하지 않고 순차적으로 학습하는 점에서 보팅,배깅과 다르다.

#         배깅
#         결론
#   예측값을 통합
#  예측  예측  예측
# 모델1 모델2 모델3

#                           부스팅
# 모델1 예측 -> 모델2 예측 -> 모델3 예측 -> 최종예측

from xgboost import XGBClassifier
xgbc = XGBClassifier(n_estimators=50, max_depth=3, random_state=20)
xgbc.fit(X_train, y_train)
y_xgbc_pred = xgbc.predict(X_test)
xgbc_acc = accuracy_score(y_test, y_xgbc_pred)
xgbc_acc
```

```python
# 과대적합 overfitting : 훈련데이터와 관련된 예측은 잘 읽지만 조금 다른 데이터는 잘 인식하지 못하는 현상

# 과소적합 underfitting : 훈련데이터가 충분하지 못해서 예측을 잘 못하는 현상
# 해결방법 : 데이터의 양을 충분히 많이 늘인다.
```

```python
# Hold-out
# X_train y_train 훈련용문제, 훈련용정답
# X_test, y_test 검증용문제, 검증용정답
X_tr  ,  X_val , y_tr , y_val = train_test_split(X_train, y_train, test_size = 0.3, shuffle=True, random_state=20)
rfc = RandomForestClassifier(max_depth=3, random_state=20)
rfc.fit(X_tr, y_tr)
y_tr_pred = rfc.predict(X_tr)
y_val_pred = rfc.predict(X_val)
tr_acc = accuracy_score(y_tr, y_tr_pred)
val_acc = accuracy_score(y_val, y_val_pred)
print(tr_acc, val_acc)
```

```python
y_test_pred = rfc.predict(X_test)
test_acc = accuracy_score(y_test, y_test_pred)
test_acc
```

```python
# k-fold
# 홀드 아웃은 학습데이터를 다시 훈련용과 검증용으로 한번만 분할하여 성능을 검증한다.
# k-hold는 홀드아웃을 여러번 반복하는 방법이다. 
# 데이터를 서로 겹치지 않게 분할하여 모두 k번 검증하므로 모델의 일반화 성능을 평가하는데 뛰어나다.

from sklearn.model_selection import KFold
kfold = KFold(n_splits=5, shuffle=True, random_state=20)
num_fold = 1
for tr_idx, val_idx in kfold.split(X_train):
  print("%s Fold----------------------------------" % num_fold)
  print("훈련: ", len(tr_idx),tr_idx[:23])
  print("검증: ", len(val_idx), val_idx[:23])
  num_fold +=1
```

```python
# 각 fold로 모델을 만들어 예측하고 그 예측의 정확도를 구해보자.

from sklearn.model_selection import KFold
kfold = KFold(n_splits=5, shuffle=True, random_state=20)

val_scores = []
num_fold=1

for tr_idx, val_idx in kfold.split(X_train,y_train):
# 훈련용데이터와 검즈용데이터를 행 인덱스를 기준으로 추출
  X_train, X_val = X_train.iloc[tr_idx,:], X_train.iloc[val_idx,:]
  y_train, y_val = y_train.iloc[tr_idx,:], y_train.iloc[val_idx,:]
#학습
  rfc = RandomForestClassifier(max_depth=5, random_state=20)
  rfc.fit(X_tr,y_tr)
#검증
  y_val_pred = rfc.predict(X_val)
  val_acc = accuracy_score(y_val, y_val_pred)
  print("%d Fold Accurancy : %.4f" %(num_fold, val_acc))
  val_scores.append(val_acc)
  num_fold +=1
```
